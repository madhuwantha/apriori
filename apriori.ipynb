```python
import numpy as np
import pandas as pd
import apriori_functions
```

```python
dataSet = pd.read_csv('Data/data_final_1_5.csv')
dataSet.head()
```

# Data Preprocessing

```python
dataSet = apriori_functions.preProcessing(dataSet)
```

# Find Frequent Itemsets

```python
C1 = apriori_functions.createC1(dataSet.to_numpy())
D = list(map(set,dataSet.to_numpy()))
L1,suppDat0 = apriori_functions.scanD(D,C1,0.5)
```

```python
freqItems,suppData= apriori_functions.apriori(dataSet.to_numpy(),minSupport=0.1)
freqItems
```

```python
np.save('output/freqItems.npy',np.array(freqItems))
```

```python
freq = np.load('output/freqItems.npy',allow_pickle=True)
freq.tolist()
```

```python
for i in range(len(freqItems)):
    print(len(freqItems[i]))
```

```python
rules= apriori_functions.generateRules(freq.tolist(),suppData, minConf=0.5)
rules
```

```python
final_Rules = pd.DataFrame(rules, columns=['ItemA', 'ItemB', 'ConfidanceA->B', 'ConfidanceB->A', 'Support(A,B)', 'Support(A)', 'Support(B)', 'Lift'])
final_Rules["ItemA"] = final_Rules.apply(lambda row: apriori_functions.convertToStringList(str(row["ItemA"])), axis=1)
final_Rules.to_csv('output/output_1_5.csv')
final_Rules
```
